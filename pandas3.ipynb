```python
#filtering of data
import pandas as pd
df = pd.read_csv("Downloads/archive (1)/globalAirQuality.csv")
df.head()
```

```python
df[df["aqi"]>100]
df[(df["aqi"]>100)&(df["temperature"]<30)]

df[df["aqi"]>100][["city","aqi"]]
```

```python
#query
df.query("aqi>100")
df.query("aqi>100 & temperature>30")[["city", "aqi"]]

aqi_val = 100
df.query("aqi > @aqi_val & temperature > 30")[["city", "aqi"]]
```
