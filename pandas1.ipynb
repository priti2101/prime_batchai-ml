```python
import pandas as pd
```

```python
info = {
    "name" : ["rahul", "harshita", "shradha"],
    "cgpa" : [9.5, 7.5, 8.5]
}
df = pd.DataFrame(info)
print(df)
```

```python
#series in pandas
s = pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#index
print(s[0])
print(s[1])
```

```python
s = pd.Series([22,34,56,67], index = ["adam", "eve", "nia", "carl"])
print(s)
```

```python
print(s[0])
print(s["carl"])
print(s.index)
```

```python
#dataframes in pandas
info = {
    "name" : ["adam", "nia", "carl"],
    "age" : [23, 45, 67],
    "gpa" : [9.5, 7.8, 8.2]
}
df = pd.DataFrame(info)
print(df, type(df))
print(df.index)
print(df.columns)
```

```python
df = pd.DataFrame([["adam", 23], ["bob", 23], ["eve", 23]], columns = ["name", "age"])
print(df)
```

```python
import numpy as np
np_arr  = np.array([[1,3,4], [4,5,6]])
df = pd.DataFrame(np_arr, columns= ["a","b","c"])
print(df)
```
